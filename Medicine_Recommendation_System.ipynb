# **Loading the Dataset**

```python
import numpy as np
import pandas as pd
df=pd.read_csv("Training.csv")
df.head()


```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
X=df.drop("prognosis",axis=1)
y=df['prognosis']
```

```python
X
```

```python
y
```

```python
le=LabelEncoder()
le.fit(y)
Y=le.transform(y)
Y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y, test_size=0.3, random_state=20)
```

# **Training Top Models**

```python


from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score,confusion_matrix

models= {
    "SVC":SVC(kernel='linear'),
    "RandomForest":RandomForestClassifier(n_estimators=100,random_state=42),
   "GradientBoosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "KNeighbors":KNeighborsClassifier(n_neighbors=5),
  "MultinomialNB":MultinomialNB()
}
for model_name,model in models.items():
  # TRAINING
  model.fit(X_train,y_train)
  # TESTING
  y_pred=model.predict(X_test)
  # ACCURACY
  accuracy=accuracy_score(y_test,y_pred)
  # Confusion matrix
  cm=confusion_matrix(y_test,y_pred)

  print(f"{model_name} has accuracy of {accuracy}")
  print(f"{model_name} Confsion matrix")
  # print(np.array2string(cm,separator=', '))
  print(cm)
```

```python

import pickle
import os

svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
acc=accuracy_score(y_test,y_pred)

models_dir = 'models'
os.makedirs(models_dir, exist_ok=True)

# Save the model in the 'models' folder
model_path = os.path.join(models_dir, 'svc_modeLl.pkl')
with open(model_path, 'wb') as file:
    pickle.dump(svc, file)
```

```python
# loading th model

with open(model_path, 'rb') as file:
    svc = pickle.load(file)

svc.predict(X_test.iloc[0].values.reshape(1,-1))
```

```python
symptoms=pd.read_csv('/content/symtoms_df.csv')
precautions=pd.read_csv('/content/precautions_df.csv')
workout=pd.read_csv('/content/workout_df.csv')
description=pd.read_csv('/content/description.csv')
medications=pd.read_csv('/content/medications.csv')
diets=pd.read_csv('/content/diets.csv')
```

```python

symptoms_dict = {col: idx for idx, col in enumerate(X)}
```

```python
disease_dict = {disease: label for disease, label in zip(y, Y)}
```

```python
def helper(dis):
    desc = description[description['Disease'] == dis]['Description']
    desc = " ".join([w for w in desc])

    pre = precautions[precautions['Disease'] == dis][['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']]
    pre = [col for col in pre.values]

    med = medications[medications['Disease'] == dis]['Medication']
    med = [med for med in med.values]

    die = diets[diets['Disease'] == dis]['Diet']
    die = [die for die in die.values]

    wrkout = workout[workout['disease'] == dis]['workout']

    return desc, pre, med, die, wrkout


def predict_disease(patient_symptoms):
  input=np.zeros(len(symptoms_dict))

  for item in patient_symptoms:
    input[symptoms_dict[item]]=1

  input=input.reshape(1,-1)
  prediction=svc.predict(input)[0]
  disease_name=list(disease_dict.keys())[list(disease_dict.values()).index(prediction)]
  return disease_name
```

```python
symptoms=input("Enter symptoms")
patient_symptoms = [symptom.strip() for symptom in symptoms.split(',')]
predicted_disease = predict_disease(patient_symptoms)
predicted_disease
desc, pre, med, die, wrkout=helper(predicted_disease)

print("=================Predicted disease============")
print(predicted_disease)
print("=================Description==================")
print(desc)
print("=================Precautions==================")
i = 1
for p_i in pre[0]:
    print(i, ": ", p_i)
    i += 1

print("=================Medications==================")
for m_i in med:
    print(i, ": ", m_i)
    i += 1

print("=================Workout==================")
for w_i in wrkout:
    print(i, ": ", w_i)
    i += 1

print("=================Diets==================")
for d_i in die:
    print(i, ": ", d_i)
    i += 1
```
